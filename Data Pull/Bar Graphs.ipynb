```python
urban_path = (r"C:\Users\yenia\Desktop\JHU_Bootcamp\Project_1\Project_one\Data Pull\urban_summary.csv")
urban_summary = pd.read_csv(urban_path)
rural_path = (r"C:\Users\yenia\Desktop\JHU_Bootcamp\Project_1\Project_one\Data Pull\rural_summary.csv")
rural_summary = pd.read_csv(rural_path)
```

```python
top_5_urban =urban_summary.nlargest(5, ["Restaurants w/ Vegan Options"])
top_5_urban
```

```python
# Generate a bar plot showing the number of restaurants with vegan options vs. top 5 cities
cities =  top_5_urban["Searched"]
vegan_rest= top_5_urban["Restaurants w/ Vegan Options"]

#Create a bar chart of the data
plt.bar(cities, vegan_rest, color = "blue")
#Create titles
plt.xlabel("City")
plt.ylabel("Number of Restaurants with Vegan Options")
plt.title("Urban Cities with the Highest Number of Vegan Friendly Restaurants")
plt.xticks(rotation = "vertical")
plt.savefig("Urban Cities with the Highest Number of Vegan Friendly Restaurants.png",bbox_inches='tight')
plt.show()
```

```python
# Generate a bar plot showing the population vs. top 5 cities 

cities =  top_5_urban["Searched"]
population = top_5_urban["Population"]

#Create a bar chart of the data
plt.bar(cities, population, color = "blue")
#Create titles
plt.xlabel("City")
plt.ylabel("Population")
plt.title("Population of Urban Cities with the Highest Number of Vegan Friendly Restaurants")
plt.xticks(rotation = "vertical")
plt.savefig("Population of Urban Cities with the Highest Number of Vegan Friendly Restaurants.png",bbox_inches='tight')
plt.show()
```
